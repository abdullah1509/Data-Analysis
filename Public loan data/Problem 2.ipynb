### JobsReported VS InitialApprovalAmount

```python
JobsReported= data['JobsReported']
InitialApprovalAmount= data['InitialApprovalAmount']
```

```python
# co-relation
correlation = JobsReported.corr(InitialApprovalAmount)
print(correlation)
```

```python
# visualization
data.groupby('JobsReported')['InitialApprovalAmount'].nunique().plot(kind='bar', figsize=(15, 10))
plt.show()
```
